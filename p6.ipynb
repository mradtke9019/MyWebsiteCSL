```python
import folium  #  map visuals
import pandas as pd  #  data manipulation
import requests
from time import sleep
```

```python
data = pd.read_csv("./foodgrading.csv")
if 'Latitude' not in data.columns and 'Longitude' not in data.columns:
    data["Latitude"] = ""
    data["Longitude"] = ""
    print("No columns")
data
```

```python
# geocode addresses for the map
latitude = []
longitude = []

for idx, e in data.iterrows():
    try:
        if len(str(data.iloc[idx]["Latitude"])) > 0:
            print("Data exists")
            latitude.append(data.iloc[idx]["Latitude"])
            longitude.append(data.iloc[idx]["Longitude"])
            continue
            
        
        address = e["ADDRESS"] + ", " + "Milwaukee" \
            + ", " + "WI"
        
        address = address.replace(" ", "+")
        request = "https://maps.googleapis.com/maps/api/geocode/json?address={}&key=AIzaSyBU828rT3AJLuU4LtZksyB2iWH63oBLCsU".format(address)
        r = requests.get(request)
        jsonInfo = r.json()
        
        locationDict = jsonInfo['results'][0]['geometry']['location']
        
        latitude.append(locationDict['lat'])
        longitude.append(locationDict['lng'] )
        
              
    except:
        print("Fail")
data["Latitude"] = latitude
data["Longitude"] = longitude
data
```

```python
data.to_csv("./cleanedData.csv", index=False)
```
